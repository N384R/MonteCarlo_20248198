```python
import numpy as np
import matplotlib.pyplot as plt
from MonteCarlo import Ising, System

def plot_property(x, y, property_name, color, ax, marker=True):
    if marker:
        ax.plot(x, y, marker='o', linestyle='-', color=color)
        ax.set_xlabel('External field (H)')
    else:
        ax.plot(x, y, linestyle='-', color=color)
        ax.set_xlabel('Temperature (T)')
    ax.set_ylabel(property_name)
```

1-1. Equilibrium simulation (N=10, steps=5000)

```python
N = 10
system = System(N, J=1)
ising = Ising(system)

ising.run_eq(T_start=10, T_end=0.1, dT=0.1, steps=5000)

temperatures = np.array(list(ising.results['magnetization'].keys()))
magnetizations = np.array(list(ising.results['magnetization'].values()))
energies = np.array(list(ising.results['energy'].values()))
specific_heats = np.array(list(ising.results['specific_heat'].values()))
susceptibilities = np.array(list(ising.results['susceptibility'].values()))

fig, ax = plt.subplots(1, 4, figsize=(20, 5))
plt.rcParams.update({'font.size': 12})
plt.subplots_adjust(wspace=0.5)
fig.set_facecolor('white')

plot_property(temperatures, magnetizations, 'Magnetization', 'black', ax[0])
plot_property(temperatures, energies, 'Energy', 'blue', ax[1])
plot_property(temperatures, specific_heats, 'Specific Heat', 'red', ax[2])
plot_property(temperatures, susceptibilities, 'Susceptibility', 'green', ax[3])
```

2-1. Non-equilibrium simulation (N=10, alpha=0.8, gamma=1)

```python
N = 10
system = System(N, J=1, gamma=1)
ising = Ising(system)
ising.run_non_eq(T=0.8, H_lmax=-5, H_rmax=5)

H_values = np.array(list(ising.results['magnetization'].keys()))
magnetizations = np.array(list(ising.results['magnetization'].values()))
energies = np.array(list(ising.results['energy'].values()))

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
plt.subplots_adjust(wspace=0.5)
plt.rcParams.update({'font.size': 12})
fig.set_facecolor('white')

plot_property(H_values, magnetizations, 'Magnetization', 'black', ax[0], marker=False)
plot_property(H_values, energies, 'Energy', 'blue', ax[1], marker=False)
```
